When analyzing a database, the Pandas library is used, which automatically determines the data type of each column in a DataFrame based on its content. However, changing the data type of the columns listed below can help speed up data processing, fix errors, and bring the data into the desired format for analysis.

```python
df_sales['Date'] = pd.to_datetime(df_sales.Date,format = '%Y-%m-%d')
df_sales['Unit_Cost'] = df_sales['Unit_Cost'].astype(float)
df_sales['Unit_Price'] = df_sales['Unit_Price'].astype(float)
df_sales['Profit'] = df_sales['Profit'].astype(float)
df_sales['Cost'] = df_sales['Cost'].astype(float)
df_sales['Revenue'] = df_sales['Revenue'].astype(float)


print((df_sales.dtypes))

```

### Annual costs, profit, and Revenue

#data Preparation
1) create new dataset containing the colums year, costs, profit,Revenue
2) Exclude 2016 date, because the 2016 data is only till 2016-07-31.
3) Aggregate the data by year column.


```python
df_annual_data = pd.DataFrame(df_sales, columns =["Year", "Cost", "Profit", "Revenue"])
#df_annual_data.head()
#Exlude year 2016
flt = (df_sales["Year"] < 2016)
df_annual_data  = df_annual_data.loc[flt, :]
df_annual_data.head()
#df_annual_data["Year"].max()
```

```python
df_annual_aggregation = df_annual_data.groupby("Year")

#Aggregating Cost, Profit and Revenue using sum
df_plot_data = df_annual_aggregation.agg("sum")
display(df_plot_data )
```

### Making a Plot: Annual Costs, Profit, and Revenue
1) Area chat 
2) Line Charts showing costs, profit and revenue by year



```python
import matplotlib.pyplot as plt
# Plotting data
year = df_plot_data.index.values
annual_cost = df_plot_data["Cost"]/1000
annual_profit = df_plot_data["Profit"]/1000
annual_revenue = df_plot_data["Revenue"]/1000

colors = {
    "annual_cost": "#FEAE73",
    "annual_profit": "#006955",
    "annual_revenue": "#DF0113"
    
}
plt.figure(figsize = (10,10))
plt.subplot(211)
plt.plot(year,annual_cost, label = "Annual Cost",color=colors["annual_cost"],linewidth =2)
plt.plot(year,annual_profit, label = "Annual Profit",color=colors["annual_profit"],linewidth =2)
plt.plot(year,annual_revenue, label = "Annual Revenue",color=colors["annual_revenue"],linewidth =2)

plt.grid()
#Update y
plt.yticks(fontsize = 12)
#update x-label

plt.ylabel("Amount in 10^3 USD",fontsize = 12)
plt.legend()
#update x-axis
plt.xticks(year,fontsize = 12)

plt.suptitle("\nAnnual Cost,Profit,and Revenue\n (in thousands USD)", fontsize =18)

plt.show

plt.subplot(212)
plt.stackplot(year, annual_cost, annual_profit,
             labels = ["Annual_cost", "Annual_Profit"],
             colors = [colors["annual_cost"], colors["annual_profit"]]
             )
plt.plot(year, annual_revenue, label = "Annual Revenue", color = colors["annual_revenue"], linewidth = 3)
plt.xlabel("Year",fontsize = 12)
plt.ylabel("Amount in 10^3 USD",fontsize = 12)
plt.xticks(year,fontsize = 12)
plt.legend()

plt.grid()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("bike_sales.csv")

# Define the order of months
months_order = ["January", "February", "March", "April", "May", "June", 
                "July", "August", "September", "October", "November", "December"]


plt.figure(figsize=(10, 8))


g = sns.catplot(
    x="Month", 
    y="Value", 
    hue="Variable",
    row="Year",
    kind="bar",
    order=months_order, 
    data=pd.melt(df, id_vars=["Year", "Month"], 
                 value_vars=["Cost", "Profit", "Revenue"], 
                 var_name="Variable", 
                 value_name="Value"),
 
)
plt.suptitle('Monthly Costs, Profit, Revenue', fontsize=18, y=1.03)

for ax in g.axes.flat:
    g.set_xticklabels(months_order, rotation=90)


plt.show()
```

```python
#2
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
df_sales = pd.read_csv("bike_sales.csv")

# Group data by year and customer gender
group_by_year_gender = df_sales.groupby(["Year", "Customer_Gender"])

# Calculate the sum of profit, cost, and revenue for each group
group_by_year_gender_total = group_by_year_gender[["Profit", "Cost", "Revenue"]].sum()

# Plot the stacked bar chart
group_by_year_gender_total.plot(kind='bar', figsize=(10, 5))

# add axis labels and a title
plt.xticks(rotation=45)

plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Sales costs, profit, and revenue by year and by gender')

# show the plot
plt.show()


```
